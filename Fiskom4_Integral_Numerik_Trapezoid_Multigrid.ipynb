```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trapezoid(f, a, b, n):
    h = (b-a)/n
    sum = 0.0
    for i in range(1, n):
        x = a+i*h
        sum = sum + f(x)
    integral = (h/2)*(f(a)+2*sum+f(b))
    return integral
```

```python
# from trapezoid import*
def f(x):
    return np.sin(x)


a = float(input("Batas bawah = "))
b = float(input("Batas atas = "))
n = int(input("Jumlah grid = "))
print({'a': a, 'b': b, 'n': n})
integral = trapezoid(f, a, b, n)

print('Integral = ', integral)
```

```python
# Menghasilkan nilai x pada grid
x_values = np.linspace(a, b, n+1)

# Menghitung nilai fungsi pada setiap titik x
y_values = f(x_values)

# Menghitung lebar setiap subinterval (delta x)
delta_x = (b - a) / n

# Menghitung integral Riemann menggunakan metode trapesium
integral_riemann = delta_x * \
    (np.sum(y_values) - 0.5 * (y_values[0] + y_values[-1]))
x1 = np.linspace(a, b, 100000)
y1 = f(x1)
# Plot fungsi
plt.plot(x1, y1, label='$f(x)=\sin x$')
plt.plot(x_values, y_values, label='Integral Trapezoid')
# Plot batang untuk menampilkan grid trapesium
for i in range(n):
    x_rect = [x_values[i], x_values[i+1],
              x_values[i+1], x_values[i], x_values[i]]
    y_rect = [0, 0, f(x_values[i+1]), f(x_values[i]), 0]
    plt.plot(x_rect, y_rect, 'k--', alpha=0.5)

# Fill area di bawah kurva
plt.fill_between(x_values, y_values, alpha=0.2, label='Integral')

# Menambahkan label dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.title('Integral Trapezoid dari $f(x)=\sin x$')
plt.legend()

# Menampilkan hasil
plt.show()

# Menampilkan hasil integral Riemann
print(
    f'Integral Trapezoid dari f(x)=sin(x) dari {a} hingga {b} dengan {n} subinterval: {integral_riemann}')
```
